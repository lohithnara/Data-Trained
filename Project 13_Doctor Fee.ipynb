```python
# experience  column requries integer values from string.
df["Experience"] = df["Experience"].str.split()
df["Experience"] = df["Experience"].str[0].astype("int")
```

```python
df
```

```python
# split the rating into integer.
df['Rating'].fillna('0%',inplace = True)
df['Rating'] = df['Rating'].str.slice(stop=-1).astype(int)
```

```python
df
```

```python
df["Qualification"]=df["Qualification"].str.split(",")
Qualification ={}
for x in df["Qualification"].values:
    for each in x:
        each = each.strip()
        if each in Qualification:
            Qualification[each]+=1
        else:
            Qualification[each]=1
```

```python
most_qua = sorted(Qualification.items(),key=lambda x:x[1],reverse=True)[:10]
final_qua =[]
for tup in most_qua:
    final_qua.append(tup[0])
for title in final_qua:
    df[title]=0
    
for x,y in zip(df["Qualification"].values,np.array([idx for idx in range(len(df))])):
    for q in x:
        q = q.strip()
        if q in final_qua:
            df[q][y] = 1
df.drop("Qualification",axis=1,inplace=True)
```

```python
df
```

```python
# get dummy values.
df= pd.get_dummies(df,columns=['Profile','Place'],prefix=['Profile','Place'])
```

```python
# dropping the value.
df.drop("Miscellaneous_Info",axis=1,inplace=True)
```

```python
df
```

```python
# shape of dataset.
df.shape
```

```python
# Splitting the data into x and y.
X = df.drop('Fees',axis=1)
y = df['Fees']
```

```python
# Importing train test split to train and test the model.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
# feature scaling
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
# Importing  SVM(SVR) algorithm to the model.
from sklearn.svm import SVR
m = SVR(gamma="scale")
m.fit(scaler.transform(X_train),y_train)
```

```python
def score(y_pred,y):
    y_pred = np.log(y_pred)
    y = np.log(y)
    return 1 - ((np.sum((y_pred-y)**2))/len(y))**1/2
# Prediction
y_pred = m.predict(scaler.transform(X_test))
score(y_pred,y_test)
```

```python
# hyperparameter tunning the model.
from sklearn.model_selection import GridSearchCV
parameters = {"C":[0.1,1,10],"kernel":["linear","rbf","poly"]}
reg = GridSearchCV(m,param_grid=parameters,n_jobs=-1,cv=5)
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(scaler.transform(X_test))
score(y_pred,y_test)
```
