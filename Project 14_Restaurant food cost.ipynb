```python
# get dummy values to the columns.
df4= pd.get_dummies(df4,columns=['TITLE','CUISINES','CITY','LOCALITY'],prefix=['TITLE','CUISINES','CITY','LOCALITY'])
```

```python
# checking the dataset.
df4
```

```python
# dropping the values.
df4.drop(['RATING','TIME','VOTES'],axis=1, inplace=True)
```

```python
# checking the dataset.
df4
```

```python
# Splitting the data into x and y.
X = df4.drop(['COST','RESTAURANT_ID'],axis=1)
y = df4['COST']
```

```python
# Importing train test split to train and test the model.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
# feature scaling
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
#Importing  SVR algorithm to the model.
from sklearn.svm import SVR
m = SVR(gamma="scale")
m.fit(scaler.transform(X_train),y_train)
```

```python
def score(y_pred,y):
    y_pred = np.log(y_pred)
    y = np.log(y)
    return 1 - ((np.sum((y_pred-y)**2))/len(y))**1/2
```

```python
# Prediction
y_pred = m.predict(scaler.transform(X_test))
score(y_pred,y_test)
```
