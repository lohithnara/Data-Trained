```python
# Getting dummy values to know the sub_columns of the Country,Dates.
def get_uniques(df1,columns):
    return {column: list(df1[column].unique()) for column in columns}
```

```python
# Finding Categorical columns.
categorical_columns = ['Country','Date']
get_uniques(df1,categorical_columns)
```

```python
# Importing Label encoder and standard scaler.
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
# Getting dates through date column.
df1['Date'].unique()
```

```python
# ordering the date.
date_ordering=sorted(df1['Date'].unique())
```

```python
# Indexing the date to know how many are present.
date_ordering.index('2020-04-26')
```

```python
# Defining the ordinal encode.
def ordinal_encode(df,column,ordering):
    df1 =df.copy()
    df1[column] = df1[column].apply(lambda x: ordering.index(x))
    return df1
```

```python
# Using one hot encode to order the date.
df1 = ordinal_encode(df1,'Date',date_ordering)
```

```python
# Label encoding the data for column type.
label_encoder = LabelEncoder()
df1['Country'] = label_encoder.fit_transform(df1['Country'])
```

```python
df1
```

```python
# Splitting the data into x and y.
x= df1.drop(['Total Deaths',],axis=1)
x
```

```python
y = df1.iloc[:,5:6]
```

```python
y
```

```python
x.shape
```

```python
y.shape
```

```python
# Importing train test split to train and test the model.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
# Importing Linear regression algorithm to the model.
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
Lr.fit(x_train,y_train)
y_pred = Lr.predict(x_test)
```

```python
# Predicted model.
y_pred
```

```python
# Finding r2 score of the data
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
# Finding the mean absolute error,mean squared and root mean squared error.
from sklearn.metrics import mean_squared_error,mean_absolute_error
print('Mean absolute error:',mean_absolute_error(y_test,y_pred))
print('Mean squared error :',mean_squared_error(y_test,y_pred))
print('Root Mean squared error:',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
# Accuracy score of the training dataset.
Lr.score(x_train,y_train)
```

```python
# importing cross-validation from sklearn.
from sklearn.model_selection import cross_val_score
```

```python
# defining cross-validation.
def get_cv_scores(model):
    scores = cross_val_score(model,
                             x_train,
                             y_train,
                             cv=5,
                             scoring='r2')
    print('CV Mean: ', np.mean(scores))
    print('\n')
```

```python
# cross-validation scores.
get_cv_scores(Lr)
```

```python
# saving the model as a pickle in the file.
from sklearn.externals import joblib
joblib.dump(Lr,'Country_wide_COVID 19_Datasets .pkl')
```
