```python
# Importing the libraries.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')
%matplotlib inline
```

```python
# Read the dataset.
df= pd.read_csv('Election')
```

```python
df
```

```python
# finding the null values.
df.isnull().sum()
```

```python
# information of the data.
df.info()
```

```python
# Getting dummy values to know the sub_columns of the territory name,Party.
def get_uniques(df,columns):
    return {column: list(df[column].unique()) for column in columns}
```

```python
# Finding Categorical columns.
categorical_columns = ['territoryName','Party']
get_uniques(df,categorical_columns)
```

```python
# get dummy values to the Party column.
pd.get_dummies(df['Party'])
```

```python
# Defining the one hot encode.
def onehot_encode(df,column):
    df = df.copy()
    dummies = pd.get_dummies(df[column])
    df = pd.concat([df,dummies], axis=1)
    df.drop(column,axis=1,inplace=True)
    return df
```

```python
# Using one hot encode to get dummies into the dataset. 
df = onehot_encode(df,'Party')
```

```python
# checking the dataframe.
df
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
# Label encoding the data for column type.
label_encoder = LabelEncoder()
df['territoryName'] = label_encoder.fit_transform(df['territoryName'])
```

```python
df
```

```python
# Splitting the data into x and y.
x= df.drop(['time'],axis=1)
x
```

```python
y = df.iloc[:,2:]
```

```python
y
```

```python
# x shape.
x.shape
```

```python
# y shape.
y.shape
```

```python
# Importing train test split to train and test the model.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
# Importing Linear regression algorithm to the model.
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
Lr.fit(x_train,y_train)
y_pred = Lr.predict(x_test)
```

```python
# Predicted model.
y_pred
```

```python
# Finding r2 score of the data
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
# Finding the mean absolute error,mean squared and root mean squared error.
from sklearn.metrics import mean_squared_error,mean_absolute_error
print('Mean absolute error:',mean_absolute_error(y_test,y_pred))
print('Mean squared error :',mean_squared_error(y_test,y_pred))
print('Root Mean squared error:',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
# Accuracy score of the training dataset.
Lr.score(x_train,y_train)
```

```python
# importing cross-validation from sklearn.
from sklearn.model_selection import cross_val_score
```

```python
# defining cross-validation.
def get_cv_scores(model):
    scores = cross_val_score(model,
                             x_train,
                             y_train,
                             cv=5,
                             scoring='r2')
    print('CV Mean: ', np.mean(scores))
    print('\n')
```

```python
# cross-validation scores.
get_cv_scores(Lr)
```
