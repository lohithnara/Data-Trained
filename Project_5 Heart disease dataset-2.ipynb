```python
df1 = df.drop(columns = ['trestbps','chol','fbs','restecg','thalach'])
df1
```

```python
# modeling

X = df1.drop('num',axis=1)
y = df1['num']
```

```python
# splitting the data into training and testing datasets
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42,shuffle=True)
regressor = LogisticRegression()
```

```python
# Using logistic regression to the model because the target variable contains as categorical values
regressor.fit(X_train,y_train)
```

```python
# predicted values of the test dataset
y_pred = regressor.predict(X_test)
```

```python
# score of the training dataset and testing dataset
regressor.score(X_test,y_test)
regressor.score(X_train,y_train)
```

```python
# Accuracy of Predicted dataset.
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred,labels=[0,1])
confusion_matrix
```

```python
# Finding the confusion matrix and classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# Finding Croos-validation score of the model
from sklearn.model_selection import cross_val_score
cross_val_score(LogisticRegression(),X,y,cv=5,scoring='accuracy').mean()
```

```python
# saving the model as a pickle in the file.
from sklearn.externals import joblib

joblib.dump(LogisticRegression,'Heart disease dataset.pkl')
```
