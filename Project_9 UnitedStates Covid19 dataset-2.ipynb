```python
# Getting dummy values to know the sub_columns of the Province_State,Country_Region,Last_Update.
def get_uniques(df,columns):
    return {column: list(df[column].unique()) for column in columns}
```

```python
# Finding Categorical columns.
categorical_columns = ['Province_State','Country_Region','Last_Update']
get_uniques(df,categorical_columns)
```

```python
# Importing Label encoder and standard scaler.
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
# Finding the dummie values of the region column.
pd.get_dummies(df['Country_Region'])
```

```python
# Getting dates through date column.
df['Last_Update'].unique()
```

```python
# ordering the date.
date_ordering=sorted(df['Last_Update'].unique())
```

```python
# Indexing the date to know how many are present.
date_ordering.index('2020-05-21 02:32:54')
```

```python
# Defining the ordinal encode.
def ordinal_encode(df,column,ordering):
    df =df.copy()
    df[column] = df[column].apply(lambda x: ordering.index(x))
    return df
```

```python
# Defining the one hot encode.
def onehot_encode(df,column):
    df = df.copy()
    dummies = pd.get_dummies(df[column])
    df = pd.concat([df,dummies], axis=1)
    df.drop(column,axis=1,inplace=True)
    return df
```

```python
# Using one hot encode to order the date.
df = ordinal_encode(df,'Last_Update',date_ordering)
```

```python
# Using one hot encode to get dummies into the dataset. 
df = onehot_encode(df,'Country_Region')
```

```python
# Checking DataFrame.
df
```

```python
# Label encoding the data for column type.
label_encoder = LabelEncoder()
df['Province_State'] = label_encoder.fit_transform(df['Province_State'])
```

```python
df
```

```python
# Splitting the data into x and y.
x= df.drop(['ISO3'],axis=1)
x
```

```python
y=df.iloc[:,5:7]
```

```python
y
```

```python
# x shape
x.shape
```

```python
# y shape
y.shape
```

```python
# Importing train test split to train and test the model.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
# Importing Linear regression algorithm to the model.
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
Lr.fit(x_train,y_train)
y_pred = Lr.predict(x_test)
```

```python
# Predicted model.
y_pred
```

```python
# Finding r2 score of the data
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
# Finding the mean absolute error,mean squared and root mean squared error.
from sklearn.metrics import mean_squared_error,mean_absolute_error
print('Mean absolute error:',mean_absolute_error(y_test,y_pred))
print('Mean squared error :',mean_squared_error(y_test,y_pred))
print('Root Mean squared error:',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
# Accuracy score of the training dataset.
Lr.score(x_train,y_train)
```

```python
# importing cross-validation from sklearn.
from sklearn.model_selection import cross_val_score
```

```python
# defining cross-validation.
def get_cv_scores(model):
    scores = cross_val_score(model,
                             x_train,
                             y_train,
                             cv=5,
                             scoring='r2')
    print('CV Mean: ', np.mean(scores))
    print('\n')
```

```python
# cross-validation scores.
get_cv_scores(Lr)
```

```python
# saving the model as a pickle in the file.
from sklearn.externals import joblib
joblib.dump(Lr,'UnitedStates Covid19 dataset.pkl')
```
